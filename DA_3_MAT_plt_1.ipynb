```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
a = np.linspace(0,10,11)
b = a**4
print(a)
print(b)

x = np.arange(0,10)
y = 2*x
print(x)
print(y)
```

```python

# create an empty canvas
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   # left,bottom,width,height
axes.plot(a,b)
plt.show()
```

```python

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   # left,bottom,width,height
axes.plot(x,y)
plt.show()
```

```python
fig,axes = plt.subplots()
axes.plot(y,x,'r') # r= red color
axes.set_xlabel('y')
axes.set_ylabel('x')
```

```python
fig,axes = plt.subplots()
axes.plot(b,a,'b') # r= red color
axes.set_xlabel('b')
axes.set_ylabel('a')
```

```python
fig , axes = plt.subplots(nrows = 2, ncols=2, figsize=(12,8))

# parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(b,a)
axes[1][0].plot(y,x)

# use l,r,t,b to stretch subplots
# use wspace,hspace to add spacing btn subplots

fig.subplots_adjust(left=None,bottom = None,right=None,top=None,wspace=0.9,hspace=0.1,)
plt.show()
```
